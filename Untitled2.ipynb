Ao analizar o arquivo e suas colunas realizamos um primeiro processo de filtro, selecionando as colunas pertinentes ao projeto.

```python
coluna_de_seleção = ['NO_MUNICIPIO_RESIDENCIA','NU_IDADE', 'TP_SEXO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_COMP1','NU_NOTA_COMP2',
                     'NU_NOTA_COMP3','NU_NOTA_COMP4', 'NU_NOTA_COMP5','NU_NOTA_REDACAO']
```

```python
dados_enem.filter(items=coluna_de_seleção)
```

```python
dados_enem=dados_enem.dropna()
```

Conforme a imagem acima, o arquivo ainda continha dados incosistentes, que foram retirados utilizando o comando "dropna" da biblioteca pandas.

```python
cidade_nota_cn=['NO_MUNICIPIO_RESIDENCIA' , 'NU_NOTA_CN']
```

```python
notacn=dados_enem.filter(items=cidade_nota_cn)
```

```python
notacn.groupby('NO_MUNICIPIO_RESIDENCIA').max()
```

Realizei um teste inicial de analise, selecionando a coluna correspondente a municipio e nota de CN, realizei o filtro e em seguida foi agrupado com indicação de visualização de nota máxima por municipio.

```python
cidade_nota_gr=['NO_MUNICIPIO_RESIDENCIA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_COMP1','NU_NOTA_COMP2',
                     'NU_NOTA_COMP3','NU_NOTA_COMP4', 'NU_NOTA_COMP5','NU_NOTA_REDACAO']
```

```python
notagr=dados_enem.filter(items=cidade_nota_gr)
```

```python
notagr.groupby('NO_MUNICIPIO_RESIDENCIA').max()
```

Realizamos o teste anterior adicionando todas as colunas de notas maximas correspondentes aos municipios. Podemos observar que a melhor nota dos municipios analizados corresponde a PIRACICABA e logo em seguida SÃO PEDRO.

```python
nota_media_geral=notagr.groupby('NO_MUNICIPIO_RESIDENCIA').mean()
```

```python
nota_media_geral
```

Agora foi realizado uma analise de notas medias por municipio, podemos perceber que o municipio de SÃO PEDRO obteve as maiores notas por média geral dos municipios analizados.
